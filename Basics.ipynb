```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
star_wars = pd.read_csv("star_wars.csv", encoding="ISO-8859-1")

star_wars = pd.DataFrame(star_wars)
star_wars.head(10)
```

```python
star_wars.columns
```

```python
star_wars['Have you seen any of the 6 films in the Star Wars franchise?'].value_counts()
```

```python
star_wars['Do you consider yourself to be a fan of the Star Wars film franchise?'].value_counts()
```

```python
star_wars[['Have you seen any of the 6 films in the Star Wars franchise?', 'Do you consider yourself to be a fan of the Star Wars film franchise?']].head()
```

```python
yes_no = {'Yes': True, 'No': False}
cols = ['Have you seen any of the 6 films in the Star Wars franchise?', 'Do you consider yourself to be a fan of the Star Wars film franchise?']

for row in cols:
    star_wars[row] = star_wars[row].map(yes_no)
    
    
star_wars[cols].tail()
```

```python
movie_mappings = {
    'Star Wars: Episode I  The Phantom Menace': True, 
    np.nan: False, 
    'Star Wars: Episode II  Attack of the Clones': True, 
    'Star Wars: Episode III  Revenge of the Sith': True, 
    'Star Wars: Episode IV  A New Hope': True, 
    'Star Wars: Episode V The Empire Strikes Back': True, 
    'Star Wars: Episode VI Return of the Jedi': True}



for row in (star_wars.columns[3:9]):
    star_wars[row] = star_wars[row].map(movie_mappings)

star_wars.columns
```

```python
star_wars = star_wars.rename(columns={"Which of the following Star Wars films have you seen? Please select all that apply.": "seen_1", "Unnamed: 4": "seen_2", "Unnamed: 5": "seen_3", "Unnamed: 6": "seen_4", "Unnamed: 7": "seen_5", "Unnamed: 8": "seen_6"})
star_wars.head()
```

```python
star_wars[star_wars.columns[9:15]] = star_wars[star_wars.columns[9:15]].astype(float)
star_wars.head()
```

```python
star_wars = star_wars.rename(columns={"Please rank the Star Wars films in order of preference with 1 being your favorite film in the franchise and 6 being your least favorite film.": "ranking_1", "Unnamed: 10": "ranking_2", "Unnamed: 11": "ranking_3", "Unnamed: 12": "ranking_4", "Unnamed: 13": "ranking_5", "Unnamed: 14": "ranking_6"})
star_wars.head()
```

Now that we've cleaned up the ranking columns, we can find the highest-ranked movie quicker. To do this, take the mean of each ranking columns using the *pandas.DataFrame.mean()* method.

```python
star_wars[star_wars.columns[9:15]].mean()
```

```python
star_wars[star_wars.columns[9:15]].mean().plot.barh()
```

## Rankings

1. **"Star Wars: Episode V The Empire Strikes Back"** with rating of 2.5.

The second most favourite movie in the franchise is columns `ranking_6`, which is the 

2. **"Star Wars: Episode VI Return of the Jedi"** with rating of 3.

These are the original movies that came out first and it seems they are rated much higher.

## Gender Segmentation

We know which movies the survey population as a whole has ranked the highest. Now let's examine how certain segments of the survey population responded. There are several columns that segment our data into two groups. Here are a few examples:

* "Do you consider yourself to be a fan of the Star Wars film franchise?" — True or False
* "Do you consider yourself to be a fan of the Star Trek franchise?" — Yes or No
* Gender — Male or Female

We can split our DataFrame based on one of the binary columns, and create two new subset of data. We will take the `Gender` column for instance:

```python
males = star_wars[star_wars['Gender'] == 'Male']
females = star_wars[star_wars['Gender'] == 'Female']
females
```

### Male & Female Views and Rankings

Segmented below we see the respective movie ranking based on gender, first graph is for female respondednts.

```python
males[males.columns[9:15]].mean().plot.bar(title='Movie rankings, Males')
plt.show()

females[females.columns[9:15]].mean().plot.bar(title='Movie rankings, Females')
```

```python
males[males.columns[3:9]].sum().plot.bar(title='Seen movies, Males')
plt.show()

females[females.columns[3:9]].sum().plot.bar(title='Seen movies, Females')
plt.show()
```

```python
males[males.columns[9:15]].mean()
```

From the above graphs, we can conclude that female population that participated in the survey gave almost identical movie rankings.
However, while **"Star Wars: Episode VI Return of the Jedi"** is the second favourite movie among the female participants, male participants ranked **"Star Wars: Episode IV A New Hope"** and **"Star Wars: Episode VI Return of the Jedi"** almost the same with a rating of 2.99 and 3 respectively.


Another takeaway from these graphs is that female respondents seen the movies less than the male population. And males watched movies 1 - 4 much more, but e rankings remained low for these 4 movies.
